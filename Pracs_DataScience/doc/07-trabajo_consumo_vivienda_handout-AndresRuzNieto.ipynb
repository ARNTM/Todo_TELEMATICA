## Primer paso: cargamos los datos
Después de importar los módulos necesarios y definir `DATA_DIRECTORY`, cargar el fichero `household_hourly_power_consumption.txt` en un DataFrame llamado `vivienda`. 

```python
# Completar aquí

import pandas as pd
from os import path
DATA_DIRECTORY = path.join('..', 'data')

vivienda = pd.read_csv(
    path.join(DATA_DIRECTORY,  'household_hourly_power_consumption.txt'),
    parse_dates = ['date_hour'],
    sep=';',
    skiprows= 13,
    index_col= 'date_hour'
)

# --------------------
vivienda
```

- sub_metering_3: termo eléctrico de ACS y un aire condicionado.

## Manipulaciones

### Añadimos una columna *sub_metering_resto*

Las columnas sub_metering_1, sub_metering_2 y sub_metering_3 miden la energía activa en tres zonas de la vivienda. Para calcular la energía activa en el resto de la vivienda, debemos substraerlas de la columna global_active_power, (despúes de multiplicar está última por 60/1000 para pasar de kW por minuto  a W por hora), según la fórmula

> (global_active_power*1000/60 - sub_metering_1 - sub_metering_2 - sub_metering_3) 

Tenéis que añadir esta columna que llamaréis sub_metering_resto al dataframe *vivienda*

```python
# Completar aquí
vivienda.insert(7, 'sub_metering_resto', ((vivienda['global_active_power'] * 1000/60) - vivienda['sub_metering_1'] - vivienda['sub_metering_2'] - vivienda['sub_metering_3']))

# --------------------
vivienda
```

## Resumimos por grupos
En esta sección usaremos `groupby` aplicado al índice del DataFrame `vivienda` para obtener distintos resúmenes del consumo energético de la vivienda.

### Perfil de potencia a lo largo del día

Queremos ver para empezar el perfil de potencia global activa medio por hora, es decir para cada hora del día (0 a 23), cuál es el valor promedio de la potencia global activa.

```python
# Completar aquí

vivienda['global_active_power'].groupby([vivienda.index.hour]).mean()

# --------------------
```

Repetir la instrucción anterior para añadir a la vez la potencia global máxima por hora, el número de datos que han entrado en el cálculo, y la potencia mínima. Guardaréis el resultado en un DataFrame  llamado `perfil_horario_vivienda`.

```python
# Completar aquí

perfil_horario_potencia = vivienda['global_active_power'].groupby([vivienda.index.hour]).mean().to_frame()

perfil_horario_potencia.rename(columns = {'global_active_power': 'potencia_media'}, inplace = True)

perfil_horario_potencia.insert(1,'potencia_maxima', vivienda['global_active_power'].groupby([vivienda.index.hour]).max().to_frame())

perfil_horario_potencia.insert(2,'potencia_minima', vivienda['global_active_power'].groupby([vivienda.index.hour]).min().to_frame())

    perfil_horario_potencia.insert(3,'num_datos', vivienda['global_active_power'].groupby([vivienda.index.hour]).count().to_frame())


# --------------------
perfil_horario_potencia
 
       
```

## Proporción de potencia correspondiente a la cocina

Empezamos por calcular la suma, por cada fila, de las columnas desde `sub_metering_1` hasta `sub_metering_resto`.

```python
# Completar aquí
sub_metering = vivienda.loc[:, ['sub_metering_1', 'sub_metering_2', 'sub_metering_3', 'sub_metering_resto']].sum(axis=1)
# --------------------
sub_metering
```

A continuación, añadimos al conjunto `vivienda` la columna calculada `prop_cocina` que contenga en porcentaje la proporción de `sub_metering_1` respecto a la suma de las columnas `sub_metering`.


```python
# Completar aquí
vivienda.insert(8, 'prop_cocina', ((vivienda['sub_metering_1']) / sub_metering) * 100)

# --------------------
vivienda
```

Podemos obtener ahora la evolución horario de la proporción correspondiente a la cocina, con los mismos indicadores que para la potencia global.

```python
# Completar aquí

perfil_horario_cocina = vivienda['prop_cocina'].groupby([vivienda.index.hour]).mean().to_frame()

perfil_horario_cocina.rename(columns = {'prop_cocina': 'prop_media'}, inplace = True)

perfil_horario_cocina.insert(1,'prop_maxima', vivienda['prop_cocina'].groupby([vivienda.index.hour]).max().to_frame())

perfil_horario_cocina.insert(2,'prop_minima', vivienda['prop_cocina'].groupby([vivienda.index.hour]).min().to_frame())

perfil_horario_cocina.insert(3,'num_datos', vivienda['prop_cocina'].groupby([vivienda.index.hour]).count().to_frame())


# --------------------
perfil_horario_cocina
```

Queremos añadir como factor de agrupamiento el día de la semana (atributo `weekday` de un objeto `datetime`). Calcular para el agrupamiento, día de la semana, hora, el valor promedio de la proporción correspondiente a la cocina y ordenarlos de mayor a menor. Weekday toma el valor 0 para Lunes y 6 para Domingo. Cuándo se hace más uso de la cocina en esta familia? La cocina se usa más los sábados a las 15:00


```python
# Completar aquí
vivienda['prop_cocina'].groupby([vivienda.index.weekday, vivienda.index.hour]).mean().to_frame().sort_values('prop_cocina',ascending=False)

# --------------------
```
